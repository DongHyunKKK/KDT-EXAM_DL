## 목표 : 숫자 / 패션 MNIST 데이터 분류 <hr>

### 1. 숫자 MNIST 데이터 분류

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import numpy as np
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', as_frame = False)
```

```python
X, y = mnist.data, mnist.target
X, y
```

```python
X.shape, y.shape
```

```python
X.dtype, y.dtype
```

```python
y = y.astype('float32')
featureTensor = torch.tensor(X, dtype = torch.float32)
targetTensor = torch.tensor(y, dtype = torch.float32)
```

```python
targetTensor = torch.tensor(targetTensor, dtype = torch.int64)
target_onehot = torch.zeros(targetTensor.shape[0], 10)

targetTensor = target_onehot.scatter_(1, targetTensor.unsqueeze(1), 1.0)
```

```python
X_train, X_test, y_train, y_test = featureTensor[:60000], featureTensor[60000:], targetTensor[:60000], targetTensor[60000:]
```

```python
X_train.dtype
```

```python
y_test.dtype
```

```python
y_test.shape
```

```python
scaled_X_train = X_train / 255
scaled_X_test = X_test / 255
```

```python
model = nn.Linear(784, 10)
```

```python
# optimizer 설정
optimizer = optim.SGD(model.parameters(), lr = 0.01)
```

```python
# 학습 -----------------------------------------------
nb_epochs = 10000
for epoch in range(nb_epochs + 1):

    # H(x) 계산
    prediction = model(X_train)
    
    # cost 계산
    cost = F.cross_entropy(prediction, y_train)

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # 20번마다 로그 출력
    if epoch % 20 == 0:
        print(f'Epoch {epoch:4d}/{nb_epochs} Cost: {cost.item():.6f}')
```
